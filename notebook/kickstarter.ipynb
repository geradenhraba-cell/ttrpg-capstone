```python
currency_cols = ['Funding Level', 'Funding Level.1', 'FundingLevelUSD']

def to_numeric_currency(x):
    if pd.isnull(x):
        return float('nan')
   
    cleaned = ''.join(c for c in str(x) if c.isdigit() or c in ['.', '-'])
    try:
        return float(cleaned)
    except:
        return float('nan')


for col in currency_cols:
    gamefound[col] = gamefound[col].apply(to_numeric_currency)
```
